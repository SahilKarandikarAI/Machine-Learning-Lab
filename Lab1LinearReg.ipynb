```python
x=[0,1,2,3,3,5,5,5,6,7,7,10]
y=[96,85,82,74,95,68,76,84,58,65,75,50]
```

```python
n=len(x)
```

```python
sumx=0
sumy=0
sumxy=0
sumx2=0
```

```python
for i in range(n):
    sumx=sumx+x[i]
    sumy=sumy+y[i]
    sumxy=sumxy+x[i]*y[i]
    sumx2=sumx2+ x[i]**2
```

```python
#Slope calculation
sumxsumy=sumx*sumy
m=(n*sumxy-sumxsumy)/(n*sumx2-(sumx**2))
```

```python
ybar=sumy/n
xbar=sumx/n
#Intercept calculation
c=ybar-m*xbar
print("Slope is: ",m)
print("Intercept is: ",c)
```

```python
tv=float(input("Enter television watch hours: "))
ypred=m*tv+c
print("The marks obtained after watching TV for ",tv ,"hrs is: ",ypred)
```
